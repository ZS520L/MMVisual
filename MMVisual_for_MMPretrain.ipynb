```python
import time
import openai
from IPython.display import display
from ipywidgets import widgets

# Initialize the API key
openai.api_key = "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"

def get_content(path_name):
    with open(path_name, 'r') as f:
        return f.read()

pipeline_button = widgets.Button(description='pipeline', layout = widgets.Layout(width='14%'))
runtime_button = widgets.Button(description='runtime', layout = widgets.Layout(width='14%'))
modules_button = widgets.Button(description='modules', layout = widgets.Layout(width='14%'))
datasets_button = widgets.Button(description='datasets', layout = widgets.Layout(width='14%'))
evaluation_button = widgets.Button(description='evaluation', layout = widgets.Layout(width='14%'))
schedule_button = widgets.Button(description='schedule', layout = widgets.Layout(width='14%'))

# Function to handle open button click
def on_pipeline_button_clicked(b):
    file_path_value = 'docs/zh_CN/advanced_guides/pipeline.md'
    with open(file_path_value, 'r') as file:
        doc_area.value = file.read()

# Function to handle open button click
def on_runtime_button_clicked(b):
    file_path_value = 'docs/zh_CN/advanced_guides/runtime.md'
    with open(file_path_value, 'r') as file:
        doc_area.value = file.read()

# Function to handle open button click
def on_modules_button_clicked(b):
    file_path_value = 'docs/zh_CN/advanced_guides/modules.md'
    with open(file_path_value, 'r') as file:
        doc_area.value = file.read()

def on_datasets_button_clicked(b):
    file_path_value = 'docs/zh_CN/advanced_guides/datasets.md'
    with open(file_path_value, 'r') as file:
        doc_area.value = file.read()

def on_evaluation_button_clicked(b):
    file_path_value = 'docs/zh_CN/advanced_guides/evaluation.md'
    with open(file_path_value, 'r') as file:
        doc_area.value = file.read()

def on_schedule_button_clicked(b):
    file_path_value = 'docs/zh_CN/advanced_guides/schedule.md'
    with open(file_path_value, 'r') as file:
        doc_area.value = file.read()

pipeline_button.on_click(on_pipeline_button_clicked)
runtime_button.on_click(on_runtime_button_clicked)
modules_button.on_click(on_modules_button_clicked)
datasets_button.on_click(on_datasets_button_clicked)
evaluation_button.on_click(on_evaluation_button_clicked)
schedule_button.on_click(on_schedule_button_clicked)

edit_area = widgets.Textarea(value='', placeholder='It is recommended to open the file rather than edit it directly',  rows=18, layout=widgets.Layout(width='49%'))
doc_area = widgets.Textarea(value=get_content('docs/zh_CN/advanced_guides/pipeline.md'),  rows=18, disabled=True, layout=widgets.Layout(width='49%'))

# Create text area widgets for user input, chat history and model answer
layout = widgets.Layout(width='99.5%')  # Define the layout
input_area = widgets.Textarea(value='', placeholder='Enter your question here',  rows=4, layout=layout)
chat_area = widgets.Textarea(value='', placeholder='Chat history',  rows=10, disabled=True, layout=layout)
answer_area = widgets.Textarea(value='', placeholder='AI answer',  rows=18, disabled=True, layout=layout)

# Create buttons for submitting the question, opening and saving files
submit_button = widgets.Button(description='Submit', layout = widgets.Layout(width='99.6%'))
open_button = widgets.Button(description='Open File', layout = widgets.Layout(width='20%'))
save_button = widgets.Button(description='Save File', layout = widgets.Layout(width='20%'))

# Create a text widget for entering the file path
file_path = widgets.Text(value='', placeholder='Enter file path here', description='File Path:', layout=widgets.Layout(width='50%'))

checkbox = widgets.Checkbox(
    value=False,
    description='增强模型 高消耗',
    disabled=False
)

is_save = widgets.Label('')

# display(checkbox)
# Function to handle button click
def on_button_clicked(b):
    # Here you need to get the AI's answer. For now, we'll just echo the user's question
    # You should replace this with the actual OpenAI API call
    if checkbox.value:
      prompt = """
你是一个聪明的AI,同时乐于助人！
用户正准备修改上面<>中的内容，但是遇到了一些困难。
请参考上述【】中的内容，回答用户下面的问题："""
          # Combine the question and the text box contents
      full_text = '<' + edit_area.value + '>' + '\n' + '【' + doc_area.value + '】' + prompt + '\n' + input_area.value
    else:
      
      full_text = input_area.value

    while True:
      try:
        answer = openai.ChatCompletion.create(
                model="gpt-3.5-turbo",
                messages=[{"role": "user", "content": full_text}]
              ).choices[0]['message']['content']
        break
      except Exception as e:
        print("An error occurred:", e)
    # chat_area.value = ''
    # Display user's question in chat history
    chat_area.value += "User: " + input_area.value + '\n'
    # Display AI's answer in chat history and in the answer area
    chat_area.value += "AI: " + answer + '\n'
    answer_area.value = answer

    # Clear the input area for the next question
    # input_area.value = ''

# Function to handle open button click
def on_open_button_clicked(b):
    file_path_value = file_path.value
    with open(file_path_value, 'r') as file:
        edit_area.value = file.read()

# Function to handle save button click
def on_save_button_clicked(b):
    file_path_value = file_path.value
    with open(file_path_value, 'w') as file:
        file.write(edit_area.value)
    is_save.value = 'Modify Saved'
    time.sleep(2)
    is_save.value = ''

# Register the click event handlers
submit_button.on_click(on_button_clicked)
open_button.on_click(on_open_button_clicked)
save_button.on_click(on_save_button_clicked)

display(widgets.HBox([widgets.Label('=================='), widgets.Label('cfg configuration tool'), widgets.Label('==================')]))
display(widgets.HBox([widgets.Label('Reference documents'), pipeline_button, runtime_button, modules_button, datasets_button, evaluation_button, schedule_button]))
display(widgets.HBox([edit_area, doc_area]))
# display()
display(widgets.HBox([file_path, open_button, save_button, is_save]))
display(widgets.HBox([widgets.Label('=================='), widgets.Label('Chat with GPT for help'), widgets.Label('=================='), checkbox, widgets.Label('增强模式下AI可以看到配置文件和参考文档')]))

# Display the widgets

display(input_area)
display(submit_button)
display(answer_area)
# display(widgets.HBox([widgets.Label('=================='), widgets.Label('History'), widgets.Label('==================')]))
# display(chat_area)


# Using HBox to arrange buttons horizontally
# button_box = widgets.HBox([submit_button, open_button, save_button])

```
